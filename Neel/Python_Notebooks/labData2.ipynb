```python
import pandas as pd 

#import 23andMe data as df 
twentyThree = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/PGC_UKB_23andMe_depression_10000.txt", sep = '\s+' )

#import genome wide data as df
genomeWide = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/PGC_UKB_depression_genome-wide.txt", sep = '\s+' )

#import CHR and BP data as df 
chromData = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/ALL-snps_hg19.txt", sep = '\s+' )

```

```python
#Add CHR and BP to both PGC_UKB_23andMe_depression_10000.txt based on the rs#.

#compare data in twentyThree and chrom
merge1 = pd.merge(twentyThree, chromData, on = "MarkerName",  how='outer', indicator=True)

#Rename Cols
merge1.rename(columns={'MarkerName':'SNP', 'LogOR':"OR"}, inplace=True)


#Export twentyThree to chrom based on MarkerName
twentyThree_Merged = merge1.loc[merge1["_merge"] == "both"].drop("_merge", axis=1)
twentyThree_Merged.to_csv("PGC_UKB_23andMe_depression_10000(Merged).csv",index = False)

#Export twentyThree data that does not have a pair in chrom data
twentyThree_Diff = merge1.loc[merge1["_merge"] == "left_only"].drop("_merge", axis=1)
twentyThree_Diff.to_csv("PGC_UKB_23andMe_depression_10000(no matching).csv",index = False)

#Export chrom data that did not have a match in twentyThree
chromData_Diff_TwentyThree = merge1.loc[merge1["_merge"] == "right_only"].drop("_merge", axis=1)
chromData_Diff_TwentyThree.rename(columns={'MarkerName':'SNP'}, inplace=True)
chromData_Diff_TwentyThree_temp = chromData_Diff_TwentyThree[["CHR","BP","SNP"]]
chromData_Diff_TwentyThree_temp.to_csv("ALL-snmp_hg19(no matching in 23andMe).csv",index = False)


```

```python
#Add CHR and BP to both PGC_UKB_depression_genome-wide.txt, based on the rs#.

#Compare data 
merge2 = pd.merge(genomeWide, chromData, on = "MarkerName",  how='outer', indicator=True)

#Rename Cols
merge2.rename(columns={'MarkerName':'SNP', 'LogOR':"OR"}, inplace=True)

#Export twentyThree to chrom based on MarkerName
genomeWide_Merged = merge2.loc[merge2["_merge"] == "both"].drop("_merge", axis=1)
genomeWide_Merged.to_csv("PGC_UKB_depression_genome-wide(Merged).csv",index = False)

#Export twentyThree data that does not have a pair in chrom data
genomeWide_Diff = merge2.loc[merge2["_merge"] == "left_only"].drop("_merge", axis=1)
genomeWide_Diff.to_csv("PGC_UKB_depression_genome-wide(no matching).csv",index = False)

#Export chrom data that did not have a match in twentyThree
chromData_Diff_genomeWide = merge2.loc[merge2["_merge"] == "right_only"].drop("_merge", axis=1)
chromData_Diff_genomeWide.rename(columns={'MarkerName':'SNP'}, inplace=True)
chromData_Diff_genomeWide_temp = chromData_Diff_genomeWide[["CHR","BP","SNP"]]
chromData_Diff_genomeWide_temp.to_csv("ALL-snmp_hg19(no matching in genome-wide).csv",index = False)
```
