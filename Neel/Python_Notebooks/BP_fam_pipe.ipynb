```python
#import libraries 
import pandas as pd 
import numpy as np 
```

```python
#import data 

dataFile = "/Users/npatel/Desktop/Dr.Chen_Lab/Data/BP_Fam/GenomTranscriptomTwinsMD_c1.fam"
dbGapFile = "/Users/npatel/Desktop/Dr.Chen_Lab/Data/BP_Fam/phs000486.v1.pht002804.v1.p1.GODOT_Sample.MULTI.txt"

#first data with (ID , ID, mother, father, sex, diagnosis). Goal to update diagnosis 
data = pd.read_csv(dataFile , sep = '\s+', names = ['SAMPID', 'Accession', 'F', 'M', 'S', 'D' ])

dbGap = pd.read_csv(dbGapFile , sep = '\s+', skiprows=11, names = ['dbGaP_Subject_ID', 'dbGaP_Sample_ID', 'BioSample', 'Accession', 'SAMPID', 'SUBJID', 'SAMPLE_USE' ])

pheno = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/Data/BP_Fam/phs000486.v1.pht002805.v1.p1.c1.GODOT_Subject_Phenotypes.DS-PD-RSC.txt", sep='\s+', skiprows=10)



```

```python
#match based on ID for Data and dbGap to get dbGap ID
merged = pd.merge(data, dbGap, on = "Accession",  how='outer', indicator=True)
```

```python
merged 

```

```python
matched_Accession = merged.loc[merged["_merge"] == "both"].drop("_merge", axis=1)
matched_Accession.rename(columns = {"SAMPID_x":"SAMPID"})
```

```python
#match based on ID for Data and dbGap to get dbGap phenomes 
merged2 = pd.merge(matched_Accession, pheno, on = "dbGaP_Subject_ID",  how='outer', indicator=True)
matched_Subject_ID = merged2.loc[merged2["_merge"] == "both"].drop("_merge", axis=1)
matched_Subject_ID
```

```python
exportFam_One = matched_Subject_ID[['SAMPID_x', 'Accession', 'F', 'M', 'S', 'MDD']].copy()

exportFam_One.to_string()
exportFam_One['SAMPID_x'] = exportFam_One['SAMPID_x'].astype(int)
exportFam_One['Accession'] = exportFam_One['Accession'].astype(int)
exportFam_One['F'] = exportFam_One['F'].astype(int)
exportFam_One['M'] = exportFam_One['M'].astype(int)
exportFam_One['S'] = exportFam_One['S'].astype(int)
exportFam_One['MDD'] = exportFam_One['MDD'].astype(int)
exportFam_One.to_csv("/Users/npatel/Desktop/Dr.Chen_Lab/Data/BP_Fam/BP_Fam_One.txt", index = False, header = None, sep = ' ')
exportFam_One
```

```python
exportFam_Two = matched_Subject_ID[['SAMPID_x', 'Accession', 'Sex', 'Age', 'MDD', 'Smoker', 'Alcohol', 'Menostat', 'BMI', 'Educat', 'Age2', 'Nweeks', 'MDD2', 'Smoker2', 'Alcohol2', 'Menostat2', 'Educat2', 'BMI2']].copy()
exportFam_Two = exportFam_Two.astype({"SAMPID_x":int, "Accession":int, "Sex":int, "Age":int, "MDD":int, "Smoker":int, "Alcohol":int, "Menostat":int, "BMI":float, "Educat":int, "Age2":int, "Nweeks":int, "MDD2":int, "Smoker2":int, "Alcohol2":int, "Menostat2":int, "Educat2":int, "BMI2":int})
exportFam_Two.to_csv("/Users/npatel/Desktop/Dr.Chen_Lab/Data/BP_Fam/BP_Fam_Pheno.txt", index = False, sep = ' ')
exportFam_Two
```
