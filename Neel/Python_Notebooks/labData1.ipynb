```python
import pandas as pd 
import dask.dataframe as dd

#Declare file locations
root = "/Users/npatel/Desktop/Dr.Chen_Lab/Neel_Data_Results/Task1-Eagle_Data"
#import data from files as dataframe
EAGLE_aggression_Public_2 = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/Data/EAGLE_aggression_Public_2", sep = '\s+' )
chrom_data = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/Genome_Data/ALL-snps_hg18.txt", sep='\s+')
missing = pd.read_csv("/Users/npatel/Desktop/Dr.Chen_Lab/Data/EAGLE_aggression_Public_2_FOUND.csv")
chrom_data = chrom_data.astype({"chromEnd": int})

missingChrom = chrom_data
EAGLE_aggression_Public_2
```

```python
#compare chrom data to Eagle data by RS#
merge2 = pd.merge(EAGLE_aggression_Public_2, chrom_data, on = "SNP",  how='outer', indicator=True)

# RSname only in Eagle data
Eagle_without_chrom = merge2.loc[merge2["_merge"] == "left_only"].drop("_merge", axis=1)
Eagle_without_chrom = Eagle_without_chrom.drop('#chrom',axis =1)
Eagle_without_chrom = Eagle_without_chrom.drop('chromEnd',axis =1)

missing_merge  = pd.merge(Eagle_without_chrom, missing, on = "SNP", how = 'outer', indicator=True)

Eagle_without_chrom.rename(columns={'#chrom':'CHR', 'chromEnd':'BP'}, inplace=True)

Eagle_without_chrom.to_csv(root + '/' + "EAGLE_aggression_Public_2_(no matching chrom).csv",index = False)
missing = missing_merge.loc[missing_merge["_merge"] == "both"].drop("_merge", axis=1)
missing.rename(columns={'#chrom':'CHR', 'chromEnd':'BP'}, inplace=True)
missing_merge
```

```python
#compare chrom data to Eagle data by RS#
merge3 = pd.merge(EAGLE_aggression_Public_2, chrom_data, on = "SNP",  how='outer', indicator=True)

# RSname only in Chrom data 
Chrom_only = merge3.loc[merge3["_merge"] == "right_only"].drop("_merge", axis=1)
Chrom_only.rename(columns={'SNP':'rsName', '#chrom':'CHR', 'chromEnd':'BP'}, inplace=True)
Chrom_only_temp = Chrom_only[["rsName", "CHR", "BP"]]
Chrom_only_temp = Chrom_only_temp.astype({"BP": int})
#Chrom_only_temp = Chrom_only_temp[["CHR", "BP", "SNP", "A1", "A2", "BETA", "P", "Direction", "HetISq", "HetChiSq", "HetDf", "HetPVal", "SAMPLESIZE", "ImpQuality"]]
Chrom_only_temp['CHR'] = Chrom_only_temp['CHR'].str.replace("chr","")
Chrom_only_temp.to_csv(root + '/' + "ALL-snps_hg18(no matching RS# in EAGLE).csv", index = False)

Chrom_only_temp
```

```python
#compare chrom data to Eagle data by RS#
merge4 = pd.merge(EAGLE_aggression_Public_2, chrom_data, on = "SNP",  how='outer', indicator=True)

# RSname in both data sets
both = merge4.loc[merge4["_merge"] == "both"].drop("_merge", axis=1)
both.rename(columns={'#chrom':'CHR', 'chromEnd':'BP'}, inplace=True)
both = both.astype({"BP": int})
both = both[["CHR", "BP", "SNP", "A1", "A2", "BETA", "P", "Direction", "HetISq", "HetChiSq", "HetDf", "HetPVal", "SAMPLESIZE", "ImpQuality"]]
both['CHR'] = both["CHR"].str.replace("chr","")
both.to_csv(root + '/' + "EAGLE_aggression_Public_2 (matched chrom data).csv",index = False)

both
```

```python
final = both.append(missing, ignore_index = True)
final = final[["CHR", "BP", "SNP", "A1", "A2", "BETA", "P", "Direction", "HetISq", "HetChiSq", "HetDf", "HetPVal", "SAMPLESIZE", "ImpQuality"]]
final['CHR'] = final['CHR'].str.replace("chr","")
final = final.sort_values(by = ['CHR'])
final.to_csv(root + '/' + "EAGLE_aggression_Public_2_FINAL_MERGED.txt", sep = '\t', index = False)

duplicates = final[final.duplicated(['SNP'], keep=False)]
duplicates = duplicates[["CHR", "BP", "SNP", "A1", "A2", "BETA", "P", "Direction", "HetISq", "HetChiSq", "HetDf", "HetPVal", "SAMPLESIZE", "ImpQuality"]]
duplicates["CHR"] = duplicates["CHR"].str.replace("chr","")
duplicates = duplicates.sort_values(by = ['CHR'])
duplicates.to_csv(root + '/' + "EAGLE_aggression_Public_2 duplicates.txt", sep = '\t', index = False)

without_duplicates = final.drop_duplicates(subset=['SNP'], keep = 'first')
without_duplicates = without_duplicates[["CHR", "BP", "SNP", "A1", "A2", "BETA", "P", "Direction", "HetISq", "HetChiSq", "HetDf", "HetPVal", "SAMPLESIZE", "ImpQuality"]]
without_duplicates["CHR"] = without_duplicates["CHR"].str.replace("chr","")
without_duplicates["CHR"] = without_duplicates["CHR"].astype(object)
without_duplicates = without_duplicates.sort_values(by = ['CHR'])
without_duplicates.to_csv( root + '/' + "EAGLE_aggression_Public_2_FINAL_MERGED(without Duplicates).txt", sep = '\t', index = False)
without_duplicates
```
